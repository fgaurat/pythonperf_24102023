```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic = pd.DataFrame(titanic[["survived","age","pclass","sex"]])
titanic.tail()
```

```python
titanic.info()
```

```python
titanic.dropna(axis=0,inplace=True)
titanic.info()
```

```python
titanic['sex'].replace(['male','female'],[0,1],inplace=True)
titanic
```

```python
y= titanic['survived']
X = titanic.drop('survived',axis=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X,y)
model.score(X,y)
```

```python
def has_survived(model,pclass,sex,age):
    X =np.array([age,pclass,sex]).reshape(1,3)
    return model.predict(X)

s = has_survived(model,3,0,47) # homme
s
```

```python
s = has_survived(model,1,1,47) # femme
s
```
